```python
import matplotlib.pyplot as plt
import numpy as np

# 数据准备
matrix_sizes = [1024, 2048, 3072, 4096]
threads = [8, 12, 16, 20]
thread_labels = ['8T', '12T', '16T', '20T']
x = np.arange(len(matrix_sizes))  # x 轴位置索引
width = 0.2

# Taskflow 和 TBB 执行时间（单位：ms）
tf_times = {
    8: [253.344, 6041.53, 22685.3, 56193.8],
    12: [196.851, 4164.98, 15341.5, 38371.3],
    16: [138.195, 3142.84, 11505.7, 28370.9],
    20: [188.019, 2627.22, 9516.32, 23690.0]
}
tbb_times = {
    8: [239.341, 6020.86, 22333.5, 55983.6],
    12: [166.212, 4112.81, 15318.0, 37303.9],
    16: [132.99, 3082.47, 11651.8, 28315.9],
    20: [112.304, 2623.74, 9515.35, 23435.8]
}

# 画图
fig, axes = plt.subplots(1, 2, figsize=(14, 6), sharey=True)

# 绘制 Taskflow 图
for i, t in enumerate(threads):
    positions = x + (i - 1.5) * width
    bars = axes[0].bar(positions, tf_times[t], width, label=thread_labels[i])
    # 标注加速比
    if t == 20:
        for j in range(len(matrix_sizes)):
            speedup = tf_times[8][j] / tf_times[20][j]
            axes[0].text(positions[j], tf_times[20][j] + 100, f"{speedup:.2f}x", ha='center', va='bottom', fontsize=9, color='blue')

axes[0].set_title('Taskflow Execution Time', fontsize=16)
axes[0].set_xticks(x)
# log y
axes[0].set_yscale('log')

axes[0].set_xticklabels([f"{n}x{n}" for n in matrix_sizes], fontsize=13)
axes[0].set_ylabel('Time (ms)', fontsize=14)
axes[0].legend(title='Threads', fontsize=11)
axes[0].grid(True, axis='y', linestyle='--', alpha=0.5)

# 绘制 TBB 图
for i, t in enumerate(threads):
    positions = x + (i - 1.5) * width
    bars = axes[1].bar(positions, tbb_times[t], width, label=thread_labels[i])
    # 标注加速比
    if t == 20:
        for j in range(len(matrix_sizes)):
            speedup = tbb_times[8][j] / tbb_times[20][j]
            axes[1].text(positions[j], tbb_times[20][j] + 100, f"{speedup:.2f}x", ha='center', va='bottom', fontsize=9, color='blue')

axes[1].set_title('TBB Execution Time', fontsize=16)
axes[1].set_xticks(x)
# log y
axes[1].set_yscale('log')
axes[1].set_xticklabels([f"{n}x{n}" for n in matrix_sizes], fontsize=13)
axes[1].legend(title='Threads', fontsize=11)
axes[1].grid(True, axis='y', linestyle='--', alpha=0.5)

# 总标题和布局
plt.suptitle("Matrix Multiplication: Execution Time vs Thread Count\nSpeedup of 20T over 8T", fontsize=18)
plt.tight_layout(rect=[0, 0, 1, 0.98])
plt.show()
```
